```python
from itertools import combinations
from scipy.stats import pearsonr

variables = {
    'VGR_valid': VGR_valid,
    'FMUL_valid': FMUL_valid,
    'apmdemo_valid': apmdemo_valid
}

for var1_name, var2_name in combinations(variables.keys(), 2):
    r, p = pearsonr(variables[var1_name], variables[var2_name])
    print(f"Correlation between {var1_name} and {var2_name}: r = {r:.3f}, p = {p:.3g}")
```
